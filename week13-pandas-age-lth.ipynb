```python
import pandas as pd
import numpy as np

index = pd.date_range('1/1/2000', periods=8)

df = pd.DataFrame(np.random.rand(8, 3), index=index, columns=list('ABC'))
df
```

```python
df.T
```

```python
df['D'] = df['A'] / df['B']
df
```

```python
df['E'] = np.sum(df, axis=1)
df
```

```python
df['F'] = np.ones(8)
df
```

```python
df1 = df.drop(['E', 'F'], axis='columns') # axis=1 이라고 적어도 된다.
df1.head()  # 괄호 ()안에 특정 숫자를 지정하지 않으면 처음 5개만 보여줌
```

```python
del df1['D']    # del로는 한 번에 하나의 열만 삭제할 수 있다.
df1.head()
```

```python
df1.columns = ['a', 'b', 'c']
df1
```

```python
df1.rename(columns={'a': '2020', 'b': '2021'}, inplace=True)
df1.tail()  # 전체 데이터 중 뒤에서 5개의 데이터를 보여준다.
```

```python
df1 = df1.sub(df1['2020'], axis='index')    # axis=0 이라고 적어도 된다.
df1.tail()
```

```python
df1 = df1.div(df1['c'], axis='index')    
df1.tail()
```

```python
df1.to_csv('test.csv')
df1
```

```python
import pandas as pd

df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

# iloc은 integer location의 약어로, 데이터 프레임의 행이나 열의 순서를 나타내는 정수로 특정 값을 추출합니다.
summer = df2.iloc[:, :5]
summer.columns = ['경수기',  '금', '은', '동', '합계']
summer = summer.sort_values('금', ascending=False)
summer 
```

```python
import pandas as pd

df = pd.read_html('https://ko.wikipedia.org/wiki/%EC%98%AC%EB%A6%BC%ED%94%BD_%EB%A9%94%EB%8B%AC_%EC%A7%91%EA%B3%84')
df2 = df[0].set_index('국가 (IOC 코드)')

# iloc은 integer location의 약어로, 데이터 프레임의 행이나 열의 순서를 나타내는 정수로 특정 값을 추출합니다.
summer = df2.iloc[:, :5]
summer.columns = ['경수기',  '금', '은', '동', '합계']
summer = summer.sort_values('금', ascending=False)
summer.to_excel('하계올림픽메달.xlsx') 
```
